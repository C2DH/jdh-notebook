## Introduction

To mark the centenary of the First World War, many of the former belligerent countries organised series of commemorations from November 2013 onwards. To our knowledge, this was the first large-scale commemoration to take place in the new media environment dominated by social networking platforms. Facebook and Twitter in particular were used to a large extent during the ceremonies, and also on an ongoing basis throughout the centenary.

We will be looking more specifically at the use of Twitter during the centenary. Set up in March 2006, Twitter is a microblogging social networking service that allows users to post short messages (originally 140 characters, increased to 280 in November 2017) on the platform twitter.com. These messages can be retweeted, in other words reposted either as they are or with a comment. A user can mention another user or users by writing their username (known as a handle) preceded by @, for example to start a discussion. Users can also insert a hashtag, a keyword preceded by the hash symbol (#). Hashtags, a feature invented by Twitter users, can be used to serve various purposes, for example to advocate an idea, express irony or share views on a subject that is being discussed more widely by other Twitter accounts. The nature of Twitter accounts can also vary considerably, ranging from institutional accounts to personal accounts and also various sorts of automated account (bots).

Unlike Facebook, Twitter has adopted a relatively open model when it comes to accessing its data. In April 2014, we were therefore able to start collecting data about the centenary of the First World War: using Twitters streaming application programming interface (API), we have thus far collected approximately five million tweets on the basis of keywords in English, German and French. The collection will continue, if possible, until the centenary of the Treaty of Versailles, signed with Germany on 28 June 1919. But the operational version of the database that we are using here covers the period from April 2014 to November 2016. This version of the database contains nearly 3.5 million tweets  1.1 million original tweets and approximately 2.4 million retweets , involving just over 840,000 users. In total the tweets contain more than 140,000 hashtags, 120,000 of which were used just ten times or less. Approximately 85% of the tweets collected are in English, 10% are in French and the remaining 5% are in other languages. There are relatively few are in German; our main hypothesis to explain this is that, despite the many exhibitions and publications, the First World War is not a focus of commemoration in Germany <cite data-cite="7009778/7UUQ8NJR"></cite>. Another possible explanation is that Twitter is less popular in Germany (where 7% of the population say that they use the platform) than in France (11%) and the United Kingdom (20%), according to a study published in January 2016 <cite data-cite="4583/FXRH8RIG"></cite>.

Compiling a collection of tweets of this nature raises considerable methodological issues. While we will not go into detail, we would refer our readers to previous publications that touch on these subjects <cite data-cite="7009778/GBFQ2FF7"></cite>.

One of the main advantages of this research project is that it sheds light on the potential interactions between commemorations and expressions of collective memory on the one hand, and the use of social media and especially Twitter on the other. To explore this interplay, we use what are known as distant reading techniques, in particular those outlined in Graphs, maps and trees by Franco Moretti <cite data-cite="7009778/TSIQRMTK"></cite>. In short, distant reading involves asking a computer to take over the task of reading for us when the sheer volume of data that needs to be processed becomes too great for close human reading of sources. Distant reading methods are quantitative, text-centred approaches. In practice, we will be using computer-based tools, such as data visualisation (including social network visualisation) and data mining, in other words a statistical approach to the text. Our research also combines this distant reading with a more closer reading of our sources <cite data-cite="4583/LUET8XQK"></cite>: we will be employing a macroscopic approach to understand the wider context of coverage of the centenary on Twitter, before gradually moving to a microscopic approach intended to shed light on two events that particularly stand out in the wider context  the cancellation of a concert by French rapper Black M, and the Franco-German commemoration of the Battle of Verdun on 29 May 2016.

#### The method to parse the file

`````python
import numpy as np
from collections import defaultdict, OrderedDict
import codecs
from datetime import datetime 
import pandas as pd
import matplotlib.pyplot as plt
`````

`````python
# parse the csv file produced by Iramuteq
def read_csv_file(file_path: str, minchi=3.841) -> [list, list,list]:    
    topics_meta = []
    topics = []
    line_cnt = 0
    topics_date = []
    
    with codecs.open(file_path, 'r', encoding='utf-8') as in_file:
        cluster = []
        cluster_date = []
        to_discard = False
        for line in in_file:
            line = line.rstrip('\n')
            if not line:
                topics.append(cluster)
                cluster = []
            else:
                if line_cnt < 2:
                    line_cnt += 1
                    continue  # skip two first lines
                else:
                    if line.startswith('"**";"classe";'): # indicates a new class
                        if len(cluster): # new classe; store res of the previous class
                            topics.append(cluster)
                            cluster = []
                        if len(cluster_date):
                            topics_date.append(cluster_date)
                            cluster_date = []
                    elif line == '"*****";"*";"*";"*";"*";"*"':  # delimiter line which starts the block to reject (small words)
                        to_discard = True
                    elif line == '"*";"*";"*";"*";"*";"*"':  # delimiter line which ends the block to reject
                        to_discard = False
                    elif line.startswith('"****";"'):  # info about the cluster
                        line = line.replace('"****";', '').replace('"****";""', '')
                        meta_data_line = line.split(';')
                        st = int(meta_data_line[0].strip('"'))
                        total = int(meta_data_line[1].strip('"'))
                        percentage = st/total
                        topics_meta.append((st, total, percentage))
                    else:
                        if to_discard:
                            continue  # skip words which appear between the delimiter lines (small words)
                        #print(line)
                        token_data = line.split(';')
                        chi2p = float(token_data[2].strip('"'))
                        chi2 = float(token_data[3].strip('"'))
                        p_value = float(token_data[-1].strip('"'))
                        token = token_data[-2].strip('"')
                        #keep the dates and the p_value
                        if token.startswith('*date_'):
                            date = token.replace("*date_", "")
                            cluster_date.append((date,chi2p ))
                        else:
                            if chi2 > minchi:  
                                cluster.append((token, chi2, p_value))
    if len(cluster):
        topics.append(cluster)
    if len(cluster_date):
        topics_date.append(cluster_date)
    return topics, topics_meta, topics_date              
`````

Figure 2 is a projection over time of the major themes in the subcorpus of French-language tweets. Each colour represents a theme. These themes, or classes, are composed of a series of tweets whose similarity was assessed as being statistically significant on the basis of co-occurrences of words. The words indicated in the caption are the most characteristic words in each class. The classes were created using the Reinert methodology (Reinert, 1983), as made possible by the software tool IRaMuTeQ (Ratinaud & Dejean, 2009). 

This figure enables us to analyse the temporality of the French-language subcorpus. Two temporal trends can be observed: while for most of the year the central focus of the collective memory of the First World War in the corpus is soldiers who died for France, the vocabulary used to pay tribute is more general during major commemorations and more specific (focusing on specific soldiers) for the rest of the year.

This dichotomy can be attributed to an initiative based on the database of those who died for France in the First World War published by the French Ministry of Defence on the website Mmoire des Hommes (http://www.memoiredeshommes.sga.defense.gouv.fr/). The database lists (originally with very sparse descriptive metadata) all the 1.3 million people officially classified as having Died for France  in other words they died because of events that took place during the war. The database was originally composed of images of the administrative documents confirming that each individual had died for France. An indexer plug-in was provided, enabling these images to be turned into computer-readable text. Several people decided to make use of this indexer and started using a specific hashtag (#1j1p) in their Twitter accounts. Each indexed file was given its own tweet describing the person in question, usually an infantry soldier <cite data-cite="7009778/GBFQ2FF7"></cite>. Classes 14, 1 and 2 represent this phenomenon.

Days with major commemorative events saw a different type of tweet: while the indexing process continued on these days, many tweets paid tribute more generally to the soldiers who died for France. On every 11 November and during the commemoration of the centenary of the Battle of Verdun in February 2016 and the centenary of the Battle of the Somme on 1 July 2016, the terms used to pay tribute to French soldiers were much more general.

There is one exception to this rule, however, and that is the commemoration by France and Germany of the Battle of Verdun in May 2016. The two most representative classes reflect a very different type of vocabulary: classes 7 (tombe, courir, cimetire  tomb, run, cemetery) and 10 (concert, black, annul  concert, black, cancelled). By classifying the tweets by theme and temporalizing these classes, the anomaly represented by the Franco-German commemoration of the centenary of the Battle of Verdun, already observed in our general temporal analysis based on the database of tweets (Figure 1), can be confirmed.

## Commemoration of the centenary of the Battle of Verdun

With commemorations of the two major battles of 1916, Verdun and the Somme, the year 2016 was a decisive year in the centenary celebrations, serving as a link between the events already held in 2014 and those to come in 2018  especially as the commemorations involved three of the main belligerents of the Western Front: France, the United Kingdom and Germany. The years 2015 and 2017 saw less activity (apart from the commemoration of the United States entry into the war). The centenary of the Battle of Verdun was commemorated in three stages: in February 2016, to mark the start of the battle, at an exclusively French ceremony focused on the armed forces; in May 2016, at a joint event organised by France and Germany; and with one-off ceremonies to mark more specific episodes in the Battle of Verdun.

The Franco-German commemoration gave rise to two scandals: an announcement by the Verdun local authority that a concert featuring French rapper Black M would be held to thank the young people from France and Germany who had helped prepare the scenography for the event; and the ceremony itself, choreographed by a German and involving these same young people wearing brightly coloured clothes and running through the military cemetery near the Douaumont Ossuary.

The Douaumont commemoration was not the first to spark controversy. On 11 November 2014, for example, a plane piloted by a representative of La Manif Pour Tous  a political movement set up during the debates and vote in 2013 on the French Act introducing same-sex marriage  flew over the site of Notre Dame de Lorette, where then French President Franois Hollande was inaugurating the Ring of Remembrance (Reichman, 2014). But the controversies over Verdun make it stand out from both the commemorations in February 2016 and those for the Battle of the Somme on 1 July 2016. The Notre Dame de Lorette incident received less media coverage on the whole.

Social media played a significant role in the two scandals associated with the commemoration of the Battle of Verdun, especially in disseminating information and spreading the scandal. Looking at Twitter in particular, we will try to understand this role and contextualise it in the broader fabric of the two scandals.

We will take inspiration from the notion of scandal as defined by Herv Rayner <cite data-cite="4583/NHWR8N7K"></cite>. We propose separating the purportedly scandalous event from the scandal itself: the existence of the former does not automatically give rise to the latter. One of the characteristics of a scandal is that it stands at the intersection of several social spaces. In the case of the Verdun centenary, determining the link between the two purportedly scandalous events (the concert by rapper Black M and the scenography of the commemoration) and the scandals provoked on Twitter and elsewhere involves asking the following question: how did the scandal find its way from the fachosphre (a term used to describe the far-right online presence in France  <cite data-cite="4583/B7B2APU6"></cite>; <cite data-cite="4583/A95TREFG"></cite>; <cite data-cite="4583/LVKYDPFK"></cite>), sometimes also known as the rinfosphre (a term used by the proponents themselves to indicate their determination to counter what they see as media bias  <cite data-cite="4583/HP47C3P3"></cite>; <cite data-cite="4583/QTP42GHG"></cite>), to more mainstream media with a wider audience?

### Chronology and nature of the two scandals

The scandal over the artist Black M is linked with an initiative taken by the local authority in Verdun. To thank young people from France and Germany for their participation in the commemorations, the town decided to organise a rap concert on the evening after the days commemorations. It submitted an application to the Mission du Centenaire  the inter-ministerial body coordinating the centenary commemorations  to secure funding. On 10 May 2016, the influential far-right website F de souche (http://www.fdesouche.com/), literally Native F, which presents itself as a press review <cite data-cite="4583/A95TREFG"></cite>, published a post <cite data-cite="4583/PZWWXSSN"></cite> on the Black M concert in which it highlighted lyrics in some of the rappers songs that were critical of France and French people. This post was followed by several others, including <cite data-cite="4583/8ZRVMNQT"></cite>, for example. The controversy grew, the Mission du Centenaire rejected the grant application and the concert was cancelled. Alpha Diallo, known as Black M, published a message pointing out that his grandfather, a Senegalese rifleman, had fought for France in the Second World War. Another controversy, which falls outside the scope of this article, then followed over whether Black Ms grandfather had really been a Senegalese rifleman.

On Sunday 29 May 2016, a French-German ceremony was held at the Douaumont Ossuary and military cemetery to commemorate the Battle of Verdun, with President Franois Hollande and German Chancellor Angela Merkel in attendance. German filmmaker Volker Schlndorff was appointed as artistic director for the event. During the ceremony, around 3,400 young French and Germans ran through Douaumont cemetery wearing brightly coloured clothes. The intention of the artistic director was to highlight the fact that the First World War had stripped the belligerent countries of their youth. Cemeteries play a different role in Germany and in France  as observed by George Mosse in his best-known book about the First World War, in which he described this distinction <cite data-cite="4583/DRCKFD3W"></cite> , and this might help explain some of the reactions provoked in France. The scenography was appreciated by those present at the ceremony  it was defended by the Bishop of Verdun, who also serves as director of the ossuary <cite data-cite="4583/BD2MXVG9"></cite>. Those who fomented the controversy mostly saw the images on a media platform: on television, an online video platform or Twitter. And many may not have actually seen the ceremony before tweeting.

## Conclusion

How does a purportedly scandalous event become a scandal? Has the answer to this question that has long been mulled over by researchers in humanities and social sciences, both now and in the past <cite data-cite="4583/S9SHDVT5"></cite>, changed with the emergence of social media?

By investigating the two scandals at the intersection of remembrance and politics that arose with the commemoration of the centenary of the Battle of Verdun in May 2016, we can see two distinct scenarios. The first, the cancellation of the Black M concert, gave rise to a complex circulation of information involving the web, online (and print) press and social media. In the case of the French-German commemorative ceremony itself, held on 29 May 2016, the specific nature of Twitter as opposed to the web and the press, as seen in the intense vibrations in the circulation of information, turned a purportedly scandalous event into a scandal in real time.

In both cases, the interplay between social media and other online media (a far-right press review and press websites) and offline media (the print version of some articles) took a scandal out of the environment in which it was created  which can crudely be referred to as the fachosphre. One of the consequences of the use of social media  alongside collaborative practices involving citizens, as observed for the database of people who died for France, for example  is the exacerbation of tensions over means of commemoration and, by extension, collective remembrance.

This exacerbation of tensions relating to remembrance and commemoration should be placed in a broader context associated with the workings of social media. Although we referred to Dominique Boullier, who sees Twitter as a sort of ideal social networking platform that enables information to circulate at high frequency (vibrations), Twitter can also be considered as a tool that lends itself very well to controversial usage, because it is a technology for both self-affirmation and social mobilisation (our translation, <cite data-cite="4583/B7B2APU6"></cite>, p. 146). The conjunction of these three elements  the circulation of information at high frequency, self-affirmation and social mobilisation  has consequences not only on the political field but also on collective memories.

The latter, as argued by Franois Hartog <cite data-cite="4583/SNMNUZW6"></cite>, are strongly linked to the present. Franois Hartog described our regime of historicity as one of presentism: with less emphasis on the future because of the declining notion of progress, the present has tended to become so prevalent that it changes our relationship with the past. History  a remote relationship with the past  becomes memory, in other words an interpretation of the past that is dependent on present issues. The two scandals that played out in relation to the commemoration of the centenary of the Battle of Verdun, together with an analysis of these scandals, should therefore prompt us to reconsider the very status of the present moment and memory. In the digital era, memory is undoubtedly deeply influenced by what has become a high-frequency present.
