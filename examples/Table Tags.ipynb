## Step 3: Write, Design & Preview

### Cell tagging

Function | Tag | Markdown | Explanation
--- | --- | --- | ---
Title | `title` | h1 (#) | This tag is used for the title of the article. 
Contributor | `contributor` | h3 (###) | One contributor per cell. Below the name, indicate the role and the affiliation.
Collaborator | `collaborator` |  | Collaborators can be mentioned like this: First Name, Name / orcid ID / institution.
Cover Image | `cover` | | Figures are counted automatically and will appear in the table of figures. (Captions will be added via a button?)
Disclaimer | `disclaimer` | | This disclaimer can be used for certain announcements. It will be color-coded in orange.
Keywords | `keywords` | | Keywords are to be separated by comma (“,”)
Abstract | `abstract` | | This cell features the abstract of an article with a max./min length of x characters.
Chapter Title | | h2 (##) | Chapter titles are always written in h2. 
Table | `table(-insert-name)-*` | | Tables need to be tagged using `table-` and then adding a unique name, ending with `-*`
Figure | `figure(-insert-name)-*` | | Figures need to be tagged using `figure-` and then adding a unique name, ending with `-*`
<s>Gallery of Figures</s>| <s>gallery(-insert-name)-*</s> | | <s>Multiple figures (one per cell) can be (each) tagged with gallery-(insert-name)-* to combine them into a gallery.</s>
Hermeneutics | `hermeneutics` |  | Methodology is always placed on the hermeneutics layer.
Hermeneutics Step | `hermeneutics-step` |  | This scrollytelling method is explained in chapter x
Narrative Step | `narrative-step` |  | This scrollytelling method is explained in chapter x
Hermeneutics, Narrative | `narrative` `hermeneutics` |  | Tagging both layers is a special use case explained in chapter x

### Hermeneutics

Use Case | Function | Tag | Markdown | Explanation
--- | --- | --- | --- | ---
1 | Hermeneutics | `hermeneutics` | | Everything related to hermeneutics should be tagged accordingly.
2 | Input on the hermeneutics layer, output on the narrative layer | `hermeneutics` `narrative` | | Sometimes it makes sense to have code input on the hermeneutics layer, but the output should be visible on the narrative layer as part of the story the author wants to tell. Tag both for these cases.
3 | Hermeneutics Step | `hermeneutics-step` | | Break long passages into a step by step manner to facilitate understanding while showing something visual. See chapter 4.3

### Captions for figures & tables

Use Case | Cell Type | Content Type | Tag | Caption 
--- | --- | --- | --- | ----
1 | Code | figure | `figure(-insert-name)-*` | The caption is directly added in the metadata of the code cell.
<s>2</s> | <s>Code</s> | <s>multiple figures</s> | <s>gallery(-insert-name)-*</s> | <s>The caption is directly added in the metadata of the code cell for the gallery and a seperate title for each picture.<s>
3 | Code | Table | `table(-insert-name)-*` | The caption is directly added in the metadata of the code cell.
4 | Markdown | Table | `table(-insert-name)-*` | The caption is written via the cell metadata.

### Folder structure in Github

File |Type | Explanation
--- | --- | ---
readme | .md file | This file contains the abstract of the article and the keywords.
requirements | .txt file | This file contains all the versions of the required libraries to function properly.
notebook | .ipynb file | This file is placed in the root folder and contains the full title of the article, separating the words with "-" inbetween
media | folder | This folder contains all media like images, videos, etc.
script | folder | This folder contains all the code and data needed for the notebook like the CSV-files.
