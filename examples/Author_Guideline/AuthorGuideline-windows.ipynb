### Step by Step

> ℹ️ For writing articles, we use [JupyterLab](https://jupyterlab.readthedocs.io/en/stable/) as it's the best tool to create and share documents that contain live code, media and text.

> ℹ️ To install Jupyter Lab on your device, we use [Docker](https://www.docker.com/). This tool helps us to set up the JDH writing environment wrapped in a lightweight execution environment called _containers_.<

**Step 1:** Download and install [Docker Desktop](https://www.docker.com/products/docker-desktop/) for Windows

- Make sure to check mark the use of WSL 2 (Windows Subsystem for Linux). It enables users to work in a Linux-based development environment and build Linux-based containers.
- It is possible that the WSL installation is incomplete. In this case follow the [link](https://aka.ms/wsl2kernel) as requested.

#### **Step 2:** Download and install [Github Desktop for Windows](https://desktop.github.com/)

> ℹ️ As a next step you need to connect a repository, a place where all your data is stored in one place. A repository (short 'repo') is a storage location for software packages. We use the code hosting platform Github which enables version control and collaboration.


- [Sign in](https://github.com/login) or [sign up](https://github.com/signup) for an account.

#### **Step 3:** Clone the [jupyter-stack](https://github.com/C2DH/journal-of-digital-history-jupyter-stack)

- In the jupyter-stack repository select _Open with Github Desktop_. It will open a window in Github Desktop to clone the repository on your local machine. Make sure it will be cloned into your _Github directory_. 

#### **Step 4:** Connect the jupyter-stack with Docker via the console using `docker-compose up`

- Open Docker, (read and) accept their service agreement. And sign in or sign up for an account.
- In Github Desktop select _Show in Explorer_. Copy the path of the folder location (Ctrl + C).
- Open the console or PowerShell, use the command `cd` and paste (Ctrl + V) the path. Press Enter to execute.
- Enter `docker-compose up` and press Enter. Wait until everything has been downloaded and executed. 

#### **Step 5:** Start the Docker container and open Jupyterlab in browser

- Expand the container, press the _container actions_ menu and select _Open with browser_
- A new browser tab will open to start Jupyter Lab and ask for the Token authentication
- Copy (Ctrl + C) the token from the console and paste it into the browser (Ctrl + V)

#### **Step 6:** Connect with Zotero and enter a API access key

> ℹ️ For citation management, we use the plugin [citation-manager](https://github.com/krassowski/jupyterlab-citation-manager). It connects Zotero libraries to the notebook and inserts the citations into the markdown cells. The citations are stored in the notebook metadata and references by an ID. It also renders a bibliography and inline citations.

> ℹ️ [Zotero](https://www.zotero.org/) is an open-source reference management software to manage bibliographic data and related research material. 

- Jupyter Lab is opened in the browser and requests to enter the Zotero API Access Key
- Navigate to [Zotero Settings](https://www.zotero.org/settings/keys/new) and select tab Feeds/API
- Create new private key, give it a description and save key
- Copy the newly generate key (select and Ctrl + C), navigate back to Jupyterlab and paste (Ctrl + V) the key
- Select 'Approve and synchronize'

#### **Step 7:** JDH template for Python or R

- In the folder view on the top left select either the JDH Python or R template
- Inside the template folder navigate to the .ipnyb file and open it
