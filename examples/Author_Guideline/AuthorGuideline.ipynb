# Guidelines for authors 

The journal makes use of **ipynb notebooks** to publish the articles. In addition to that, there is a couple of rules to *format content* and *visualize datasets*. The article shall be written in English (and thus use English style typography).

## Quick guide

1. Get [Jupyter Notebook](https://jupyter.org/index.html).
2. Activate [nbextensions](https://jupyter-contrib-nbextensions.readthedocs.io/en/latest/install.html) “collapsible headings” and “table of contents”.
3. Connect [cite2c](https://github.com/takluyver/cite2c) of nbextension with your Zotero account (and in there join the [JDH group](https://www.zotero.org/groups/2784584/jdh)).
4. Create a Github repository based on the appropriate template for [R kernel](https://github.com/C2DH/template_repo_JDH_R) or [Python kernel](https://github.com/C2DH/template_repo_JDH)
5. Use the [Jupyter Notebook Template for Python](https://github.com/C2DH/template_repo_JDH/blob/main/author_guideline_template.ipynb) or the [Jupyter Notebook Template for R](https://github.com/C2DH/template_repo_JDH_R/blob/master/author_guideline_template.ipynb).
6. Start writing your article, tag the beginning cells according to this guide and tag all hermeneutic cells.
7. Commit your notebook to Github and view the rendered article.
8. Submit your notebook for peer review.
9. View your published article.

## Writing: structuring the notebook

### Cell tagging

In order to render the notebook on the frontend, we are working with tags to mark cells with special content like figures, tables or the  hermeneutics layer. 
Narrative cells are generally not tagged, with one exception that is explained in chapter 4.2.1.2.

Here's an overview of the most important tags:

Function | Tag | Markdown | Explanation
--- | --- | --- | ---
Title | `title` | h1 (#) | This tag is used for the title of the article. 
Contributor | `contributor` | h3 (###) | One contributor per cell. Below the name, indicate the role and the affiliation.
Collaborator | `collaborator` |  | Collaborators can be mentioned like this: First Name, Name / orcid ID / institution.
Cover Image | `cover` | | Figures are counted automatically and will appear in the table of figures. (Captions will be added via a button?)
Disclaimer | `disclaimer` | | This disclaimer can be used for certain announcements. It will be color-coded in orange.
Keywords | `keywords` | | Keywords are to be separated by comma (“,”)
Abstract | `abstract` | | This cell features the abstract of an article with a max./min length of x characters.
Chapter Title | | h2 (##) | Chapter titles are always written in h2. 
Table | `table-1` | | Tables need to be tagged "table-(number). The number needs to be incremented **MANUALLY**.
Figure | `figure-1` | | Figures need to be tagged with "figure-(number)". The number needs to incremented **MANUALLY**. Captions can be added, see chapter 1.3.5
Gallery of Figures | `gallery-1` | | Multiple figures (one per cell) can be (each) tagged with "gallery-(number)" to combine them into a gallery. For more details see chapter 1.3.6.2
Hermeneutics | `hermeneutics` |  | Methodology is always placed on the hermeneutics layer.
Hermeneutics Step | `hermeneutics-step` |  | This scrollytelling method is explained in chapter 1.3.4
Narrative Step | `narrative-step` |  | This scrollytelling method is explained in chapter 1.3.4
Hermeneutics, Narrative | `narrative` `hermeneutics` |  | Tagging both layers is a special use case explained in chapter 1.3.3.2

For the header section we use the tags `title`, `contributor`, `cover`, `disclaimer` (if needed), `keywords` and `abstract` to show the article in a certain structure.

After the header section
- the first title is always `introduction`
- the first cell after that should be a `narrative` cell in markdown. 

In order to tag cells, enable the tagging function via View > Cell Toolbar > Tags.

### Hermeneutics

The articles in our journal have a unique layer approach which is also reflected in the way the notebooks are written.

Use Case | Function | Tag | Markdown | Explanation
--- | --- | --- | --- | ---
1 | Hermeneutics | `hermeneutics` | | Everything related to hermeneutics should be tagged accordingly.
2 | Input on the hermeneutics layer, output on the narrative layer | `hermeneutics` `narrative` | | Sometimes it makes sense to have code input on the hermeneutics layer, but the output should be visible on the narrative layer as part of the story the author wants to tell. Tag both for these cases.

### Captions for figures & tables

Figures and tables are a bit special. Depending on the type of cell there are the following use cases and guides on writing a caption:

Use Case | Cell Type | Content Type | Tag | Caption 
--- | --- | --- | --- | ----
1 | Code | figure | `figure-1` | The caption is directly added in the metadata of the code cell.
2 | Code | figure-1, figure-2 | `figure-1` | The caption is directly added in the metadata of the code cell for each figure. The cell caption is written via the cell metadata.
3 | Code | Table | `table-1` | The caption is directly added in the metadata of the code cell.
4 | Markdown | Table | `table-1` | The caption is written via the cell metadata.

### Use Cases

#### Use case 1: figure in code cell

Figures like images for example are tagged with "figure-(number)". Figures need to be **MANUALLY** incremented.
In order to add a caption (title and copyright), we use **metadata** which will be visible on the narrative layer:
```
    "tags": [figure-1],
    "jdh": {
            "object": {
            "source": [
                "Image from page 10 of \Book Title\ (1920)", 
                "Internet archive book images, no known copyright restrictions"
            ]
        }
    }
}
```

To include **videos**, use the tag "figure-(number)" as well. Here's an example for Vimeo:

```
from IPython.display import VimeoVideo

metadata={
    "jdh": {
        "module": "object",
        "object": {
            "type":"image",
            "source": [
                "figure (number): caption"
            ]
        }
    }
}

# Vimeo
display(VimeoVideo('ID','100%','height'), 
        metadata=metadata)
```

#### Use case 2: gallery

We follow a cell-centric approach, meaning **one** cell should only contain **one** figure (and thus only **one** caption). However, we offer a gallery view to showcase multiple pictures in an appealing way.
