# Guidelines for Authors {.my-5}

The journal makes use of **ipynb notebooks** to publish the articles.
In addition to that, there is a couple of rules to *format content* and *visualize datasets*, similar to what [jupyter book](https://jupyterbook.org/intro.html) does.
In practice, the journal relies heavily on **cell metadata** and **cell tags** to correctly display its structure and performing its magic: transforming a linear structure, the notebook, into a rich structured text.

- Refer to [this](https://www.datacamp.com/community/tutorials/tutorial-jupyter-notebook) or [this tutorial](https://jupyter-notebook-beginner-guide.readthedocs.io/en/latest/) for a quick guide
- [Here](https://medium.com/game-of-data/12-things-to-know-about-jupyter-notebook-markdown-3f6cef811707) is a refresher on markdown 
- and [here](https://towardsdatascience.com/jypyter-notebook-shortcuts-bf0101a98330) you’ll find some useful shortcuts. 

The first layer readers will encounter is the **narrative layer**. It features the story the author wants to tell. 

In specific instances, the author wants to reveal more of the methodology used behind certain aspects or figures. This is accomplished by placing that information on the **hermeneutics layer**. Whenever there’s more to explore, the interface indicates to the reader that there’s a connection between the narrative and hermeneutics layer. We will explore later how this can be archived visually.
Finally, the readers can jump into the **data layer** by downloading the raw data (or a subset) and exploring it themselves.

## QuickGuide

1. Get [Jupyter Notebook](https://jupyter.org/index.html).
2. Activate [nbextensions](https://jupyter-contrib-nbextensions.readthedocs.io/en/latest/install.html) “collapsible headings” and “table of contents”.
3. Connect [cite2c](https://github.com/takluyver/cite2c) of nbextension with your Zotero account (and in there join the [JDH group](https://www.zotero.org/groups/2784584/jdh)).
4. Create a [Github](https://github.com/) repository.
5. Download the [Jupyter Notebook Template](https://github.com/C2DH/jdh-notebook/blob/master/author_guideline_template.ipynb).
6. Start writing your article, tag the beginning cells according to this guide and tag all hermeneutic cells.
7. Commit your notebook to Github and view the rendered article.
8. Submit your notebook for peer review.
9. View your published article.

## Structuring the Jupyter Notebook

While the rendered article can reveal more information to the reader when needed, the notebook remains linear. 

In the next section we will introduce three ways to help structure the notebook in such a way that the author can keep track of things. 

### Cell Tagging

As a first measure we want the authors to tag the cells. Here’s an overview of the various tags and a few markdown guidelines for the beginning of the notebook:

Function | Tag | Markdown | Explanation
--- | --- | --- | ---
Title | title | h1 (#) | This tag is used for the title of the article. 
Contributors | contributors | h3 (###) | List the contributors. Below the name, indicate the role and the affiliation.
Cover Image | cover | | Figures are counted automatically and will appear in the table of figures. (Captions will be added via a button?)
Disclaimer | disclaimer | | This disclaimer can be used for certain announcements. It will be color-coded in orange.
Keywords | keywords | | Keywords are to be separated by comma (“,”)
Abstract | abstract | | This cell features the abstract of an article with a max./min length of x characters.
Chapter Title | | h2 (##) | Chapter titles are always written in h2. 
Table | `table-1` | | For tables, **individually** tag each table in the order like this: table-1, table-2, table-3.
Figure | `figure-1` | fig-1a, fig-1b | Figures need to be tagged with "figure-(number)". Captions are added manually.

### Figures & Tables

If a cell contains 2 figures in one, we propose the following cell-centric approach:
- tag the cell with figure-1
- write `fig-1a` and `fig-1b` within the cell to show them side by side on the narrative and add the corresponding caption
- modify the cell metadata to add the appropriate caption 
```
{
  "tags": ["figure-1a et 1b"],
  "jdh": {
    "object": {
      "source": ["First Figure, a and b."]
    }
  }
}
```

<code>'figure 1a: First Figure'
'figure 1b: Second Figure'</code>

## Hermeneutics

A special emphasis needs to be put on the cell tagging of content on the hermeneutics layer. For this we have two possibilities and a third option as a special input-output case. 

Function | Tag | Markdown | Explanation
--- | --- | --- | ---
Hermeneutics | hermeneutics | | Everything related to hermeneutics should be tagged accordingly.
Hermeneutics Step | hermeneutics-step | | Explanations of methodology can become very long and complicated. In order to facilitate understanding, long passages should be broken down into steps for readers to follow along more easily.
Input on the hermeneutics layer, output on the narrative layer | hermeneutics narrative | | Sometimes it makes sense to have code input on the hermeneutics layer, but the output should be visible on the narrative layer as part of the story the author wants to tell. Tag both for these cases.

## Library cite2c (for Anaconda)

1. After nbextension is installed (see here), open Anaconda prompt.
2. Check version with the following commands:

<code>anaconda --version
python --version</code>



3. Install cite2c extension.

<code>python -m pip install cite2c</code>

4. In Jupyter, check in the Nbextensions tab if cite2c is activated.
5. Open a notebook and select the button “Cite” in the toolbar. You will be asked to sign into your Zotero account. 
6. Make sure you are part of the “JournalofDigitalHistory” group in Zotero (if not request to join). 
7. If you want to add a citation to your article, simply select the “Cite” button. Type the keywords of the citation you are looking for. Press the “Insert” button to add the citation to your notebook cell. Execute the cell.
8. If you want to show your entire biography in the last cell of your notebook, click the “Insert bibliography” button in the toolbar. Execute the cell to see the result.

File |Type | Explanation

notebook | .ipynb file | This file is placed in the root folder and contains the full title of the article, separating the words with "-" inbetween
media | folder | This folder contains all media like images, videos, etc.
