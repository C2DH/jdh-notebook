# Example of area

https://vega.github.io/vega-lite/docs/area.html

## With url

```python
import pandas as pd
df_stream = pd.read_json("data/stream.json")
df_stream.head()
```

> Need to transform 
> false to False
> null to None

```python
from vega import VegaLite
VegaLite(
{
  "$schema": "https://vega.github.io/schema/vega-lite/v4.json",
  "data": {"url": "data/stream.json"},
  "mark": "area",
   "encoding": {
          "x": { 
              "field": "date", "type": "temporal", "timeUnit": "yearmonth",
              "axis": {"domain": False, "format": "%Y", "tickSize": 0}
              #"format": "YYYY-MM-DD",
              
             
          },
          "y": { 
              "aggregate": "sum", "field": "count", "type": "quantitative",
              "axis": None,
              "stack": "center"
          },
          "color": { "field": "series", "scale": { "scheme": "category20b" } }
        },
}
)
```

## With datas

```python
from pathlib import Path
txt = Path('data/stream.json').read_text()
```

```python
from vega import VegaLite
VegaLite(
{
  "$schema": "https://vega.github.io/schema/vega-lite/v4.json",
  "data":  {       
    "values": txt
  },
  "mark": "area",
  "encoding": {
          "x": { 
              "field": "date", "type": "temporal", "timeUnit": "yearmonth",
              "axis": {"domain": False, "format": "%Y", "tickSize": 0}                           
          },
          "y": { 
              "aggregate": "sum", "field": "count", "type": "quantitative",
              "axis": None,
              "stack": "center"
          },
          "color": { "field": "series", "scale": { "scheme": "category20b" } }
        },
}
)
```

# Example of point

Point mark represents each data point with a symbol. Point marks are commonly used in visualizations like scatterplots.
https://vega.github.io/vega-lite/docs/point.html

```python
import pandas as pd
df = pd.read_json('data/cars.json')
df.head()
```

```python
from vega import VegaLite
VegaLite({
  "mark": "point",
  "encoding": {
    "y": {"type": "quantitative","field": "Acceleration"},
    "x": {"type": "quantitative","field": "Horsepower"}
  }
},df)
```
